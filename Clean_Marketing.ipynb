```python
import pandas as pd
import numpy as np
path = r"C:\Project\Marketing_messy\marketing_campaign_data_messy.csv"
df = pd.read_csv(path)
df
```

```python
df.isnull().sum()
```

```python
#Handling header
df.columns = df.columns.str.lower().str.strip()
df
```

```python
#Transforming date_time type
df['start_date'] = pd.to_datetime(df['start_date'],format="mixed", errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'],format='mixed', errors='coerce')
df
```

```python
df['channel'].unique()
```

```python
#handling channel column
df['channel'] = df['channel'].replace(
    {
        'E-mail':'Email',
        'Gogle': 'Google Ads',
        'Tik_Tok':'TikTok',
        'Facebok':'Facebook',
        'Insta_gram':'Instagram',
        'N/A': np.nan
        

    }
)
df['channel'].unique()
```

```python
#Handling 2 clicks coloumn
df = df.loc[:,~df.columns.duplicated()]
df
```

```python
#Formatting Spend column
df['spend'] = df['spend'].str.strip().str.replace(r'[^\d.-]','',regex=True)
df['spend'] = pd.to_numeric(df['spend'])
df
```

```python
#Hanling active column

df['active'] = df['active'].str.replace(
    {
        'Y':'true',
        '0':'false',
        'No':'false',
        'True':'true',
        'Yes':'true',
        '1':'true',
        'False':'false',
        'truees':'true'
    }
)
df['active'].unique()
```

```python
day_differences = df['start_date'] > df['end_date']
print(df.loc[day_differences,['campaign_id','start_date','end_date']].all)

```

```python
#assuming the end_date is start_date + 30 days
df.loc[day_differences,'end_date'] = df.loc[day_differences,'start_date'] + pd.Timedelta(days=30)
print(df.loc[day_differences,['campaign_id','start_date','end_date']].all)
```

```python
Q1 = df['spend'].quantile(0.25)
Q3 = df['spend'].quantile(0.75)
IQR = Q3 -Q1
upper_limit = Q3 + 3 * IQR
outlier = df['spend'] > upper_limit
print(df.loc[outlier,['campaign_id','spend']].all)

```

```python
#Outlier = upper_limit
df.loc[outlier,['spend']] = upper_limit
print(df.loc[outlier,['campaign_id','spend']].all) 
```
